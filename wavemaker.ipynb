```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
```

```python
'''

Code for the solution to the standard wavemaker problem

'''

def phi(x_vect,z_vect,t_vect,k_vect,f,h,g):
    phi = np.zeros((len(z_vect),len(x_vect),len(t_vect)))
    for k_n in k_vect:
        C_n = C(h,k_n)
        for z in range(len(z_vect)): 
            Z_n = Z(z_vect[z],h,k_n)
            for x in range(len(x_vect)):
                X_n = X(x_vect[x],k_n)
                for t in range(len(t_vect)):
                    T_n = T(t_vect[t],h,g,k_n,f)
                    phi[z,x,t] += C_n * Z_n * X_n * T_n
    return phi
        
def C(h,k):
    C = -1*(2/k) * np.sin(k*h) / (np.cos(k*h) * np.sin(k*h) + k*h)
    return C
    
def Z(z,h,k):
    Z = np.cos(k * (z+h))
    return Z

def X(x,k):
    X = np.e**(-1 * k * x)
    return X

def T(t,h,g,k,f):
    T = 1j * np.sqrt(-1 * g * k * np.tanh(h * k)) * np.e**(-1j * t * np.sqrt(-1 * g * k * np.tanh(h * k)))
    return T
```

```python
'''

Run the model

'''


# set nondimensional height
h = 1

# make vector of nondimensional vertical positions
num_z = 20
z_vect = np.linspace(0,-1,num_z+1)
dz = 1/num_z

# make vector of nondimensional horizontal positions
num_x = 40
x_max = 2
dx = x_max/num_x
x_vect = np.linspace(0,x_max,num_x+1)

# make vector of nondimensional wavenumbers
k_vect = np.linspace(0.1,2,100)

# make vector of times
num_t = 100
t_max = 20
dt = t_max/num_t 
t_vect = np.linspace(0,t_max,num_t+1)

# set the forcing function (does this need to be scaled by height?)
def forcing(t):
    # force with a sinusoid for half the model run duration
    #if t < 2*np.pi/1.5:
    x = 0.1 - 0.1*np.sin(1.5*t)
    #else:
    #    x = 0.1 + 0.1*np.sin(2*np.pi)
    return x

# solve the wavemaker problem for velocity potential
velocity_potential = phi(x_vect,z_vect,t_vect,k_vect,forcing,h,-9.8)
```

```python
'''

Plot the result

'''

# set a few parameters
g = -9.8
rho_w = 1000

# get velocity and displacement in each direction for each time step
vertical_velocity = np.zeros((len(z_vect),len(x_vect),len(t_vect)))
horizontal_velocity = np.zeros((len(z_vect),len(x_vect),len(t_vect)))
vertical_displacement = np.zeros((len(z_vect),len(x_vect),len(t_vect)))
horizontal_displacement = np.zeros((len(z_vect),len(x_vect),len(t_vect)))
water_pressure = np.zeros((len(z_vect),len(x_vect),len(t_vect)))

forcing_displacement = []
for t in range(len(t_vect)):
    vertical_velocity[:,:,t] = np.gradient(velocity_potential[:,:,t],dz,axis=0)
    horizontal_velocity[:,:,t] = np.gradient(velocity_potential[:,:,t],dx,axis=1)
    vertical_displacement[:,:,t] = vertical_velocity[:,:,t] * dt
    horizontal_displacement[:,:,t] = horizontal_velocity[:,:,t] * dt
    water_pressure[:,:,t] = vertical_displacement[:,:,t]*rho_w*g
    forcing_displacement.append(forcing(t_vect[t]))
    
# plot velocity field
fig,ax = plt.subplots(figsize=(20,10))
cf = ax.imshow(water_pressure[:,:,1],origin="upper",extent = [x_vect[0],x_vect[-1],z_vect[-1],z_vect[0]],
               vmin=np.min(water_pressure[:,:,1]),vmax=np.max(water_pressure[:,:,1])/2)
fig.colorbar(cf)

def update_quiver(t):
    ax.clear()
    cf = ax.imshow(water_pressure[:,:,t],origin="upper",extent = [forcing_displacement[t] + x_vect[0],
              forcing_displacement[t] + x_vect[-1],z_vect[-1],z_vect[0]],
              vmin=np.min(water_pressure[:,:,1]),vmax=np.max(water_pressure[:,:,1])/2,alpha=0.5)
    ax.quiver(forcing_displacement[t] + x_vect,z_vect,horizontal_displacement[:,:,t],vertical_displacement[:,:,t])
    ax.vlines(forcing_displacement[t],-1,0,linewidth=2,colors='r',linestyles='-')
    ax.set_xlim(0,2)
    ax.set_xlabel("Nondimensional position (x/h)")
    ax.set_ylabel("Nondimensional depth (z/h)")
    ax.set_title("T = "+str(np.round(t_vect[t],2))+"s")
    
anim = animation.FuncAnimation(fig, update_quiver, frames=len(t_vect))
anim.save('flipped_sinusoidal_forcing_displacements.gif')
```
